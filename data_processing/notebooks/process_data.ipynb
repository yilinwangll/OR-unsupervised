```python
# 'NLP4LP-EquivaMap' 
# resocratic-29k-sample-3k'
# 'resocratic-29k-sample-3k-part2'
# 筛选和测试集相当的
import json

with open('/home/work/mllm_datas/yilin/code/OR-SR1/datasets/OR-Model-Eval/stage1-RL-checkpoint-878/sirl-eval-results/SIRL-OptMATH_166.pass1/executed.jsonl', 'r') as f:
    data_industry = [json.loads(line) for line in f if line.strip()]

data_industry_wrong = []
count = 0
for d in data_industry:
    sol = d.get('execution_best_solution')
    ans = d.get('en_answer')
    if sol is None or ans is None:
        data_industry_wrong.append(d)
        continue
    try:
        sol_val = float(sol)
        ans_val = float(ans)
    except (ValueError, TypeError):
        data_industry_wrong.append(d)
        continue
    rel_err = abs(sol_val - ans_val) / abs(ans_val) if ans_val != 0 else abs(sol_val)
    if rel_err <= 1e-2:
        
        count += 1
    else:
        data_industry_wrong.append(d)

print(f"Matched entries (rel_err ≤ 1e-4): {count}")
```

```python
def question2prompt(question_text: str) -> str:
    template = r"""
        You are tasked with solving an operations research problem through a structured three-step process:  
        1. Reasoning: Understand and analyze the problem.  
        2. Modeling: Formulate a precise mathematical model.  
        3. Implementation: Translate the model into executable Python code using `coptpy`.

        Please respond in the following exact format:

        <thinking>
        Your step-by-step reasoning and interpretation of the problem here.
        </thinking>
        <model>
        Your precise mathematical model formulation here, including decision variables, objective function, and constraints.
        </model>
        <code>
        Your complete and executable Python code using `coptpy` that implements the above model.
        </code>

        Question: {Question}
        """
    prompt = template.replace("{Question}", question_text.strip()).strip()
    
    return prompt
```
